## Binary Encoding.

```python
# Binary categorical columns(2, unique values)
binary_cols = [
    'gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn'
]

# Map Yes/No and Male/Female to 0/1

df[binary_cols]=df[binary_cols].replace({
    'Yes': 1, 'No':0,
    "Male":1, "Female":0
})
```

```python
df.head()
```

## One Hot Encoding

```python
df.describe(include='all')
```

```python
# Categorical columns with > 2 unique values

multi_cat_cols=[
    'MultipleLines','InternetService', 'OnlineSecurity', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
    'StreamingMovies', 'Contract', 'PaymentMethod'
]
# One-Hot Encoding
df = pd.get_dummies(df,columns=multi_cat_cols, drop_first=True)
```

```python
df.head(n=10)
```

```python
df.shape
```

## Cleaning

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'], errors='coerce') 
# this line all string values like '100' convert to 100 simple
```

```python
df=df.drop('customerID', axis=1)
```

```python
bool_cols=df.select_dtypes(include='bool').columns
```

```python
bool_cols
```

```python
for col in bool_cols:

    df[col]=df[col].astype(int)
```

```python
df.info()
```

```python
corr_matrix=df.corr(numeric_only=True)
churn_corr=corr_matrix['Churn'].sort_values(ascending=False)
print(churn_corr)
```

```python
import seaborn as sns

# Compute correlation matrix only for numric columns
corr_matrix=df.corr(numeric_only=True)

# Focus on correlation with churn
churn_corr=corr_matrix[['Churn']].sort_values(by='Churn', ascending=False)

#Plot heatmap

plt.figure(figsize=(4, 12))
sns.heatmap(churn_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

plt.title("Correlation of Feature With Churn")
plt.show()
```
